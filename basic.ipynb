```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Define the range of x values
x = np.linspace(-5, 5, 100)

# Calculate the corresponding y values using exp(x)
y = np.exp(x)

# Create the plot
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('exp(x)')
plt.title('Plot of exp(x)')
plt.grid(True)
plt.show()

```
